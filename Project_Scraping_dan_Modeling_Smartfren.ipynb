```python
from google_play_scraper import Sort, reviews_all
import csv
import pandas as pd
```

```python
# Mengambil ulasan dari site Smartfren
scrapreview = reviews_all(
    'com.smartfren',
    lang='id',
    country='id',
    sort=Sort.MOST_RELEVANT,
    count=150000
)

# Membatasi ulasan jika diperlukan (mengikuti template dosen)
scrapreview_limited = scrapreview[:150000]

# Membuat DataFrame untuk mengintip struktur data ulasan yang didapat
dataView = pd.DataFrame(scrapreview_limited)
dataView.head()
```

```python
# 1. Membuat file baru bernama hasil_scraping_smartfren.csv
with open('hasil_scraping_smartfren.csv', mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)

    # 2. Membuat nama kolom di dalam file (hanya kolom komentar dan rating)
    writer.writerow(['content', 'score'])

    # 3. Memasukkan semua ulasan Smartfren yang sudah ditarik ke dalam file
    for review in scrapreview_limited:
        writer.writerow([
            review['content'],
            review.get('score', "")
        ])

print("Data telah disimpan ke hasil_scraping_smartfren.csv")
```
